```python
def square(L=1):
    vertex = np.asarray([
        [0,0],
        [L,0],
        [L,L],
        [0,L],
        [0,0]
    ])
    return vertex.T

a = square(5)
print(a)
plt.figure(figsize=(5,5))
plt.plot(*a, 'o-')
```

```python
def square(L=1, c=(0,0)):
    vertex = np.asarray([
        [0,0],
        [L,0],
        [L,L],
        [0,L],
        [0,0]
    ])
    return (vertex - c - L/2).T

plt.figure(figsize=(5,5))
for i in range(5):
    plt.plot(*square(i,(0,0)),'o-')
```

```python
# https://es.wikipedia.org/wiki/Matriz_de_rotaci%C3%B3n
def square(L=1, c=(0,0), angle=0):
    angle = np.pi*angle/180 
    R = np.asarray([
        [np.cos(angle), -np.sin(angle)],
        [np.sin(angle), np.cos(angle)]
    ])
    vertex = np.asarray([
        [0,0],
        [L,0],
        [L,L],
        [0,L],
        [0,0]
    ])
    return np.dot((vertex - c - L/2),R).T

plt.figure(figsize=(5,5))
plt.plot(*square(i,angle=45),'o-')
```

```python
plt.figure(figsize=(5,5))

with plt.xkcd():
    for angle in np.linspace(0,90,10):
        plt.plot(*square(5, angle=angle),'o-')
```
