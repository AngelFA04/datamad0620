```python
import numpy as np
import seaborn as sns
np.random.seed(0)
sns.set()
```

```python
import matplotlib.pyplot as plt
uniform_data = np.random.random((10, 12))

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.heatmap(uniform_data)
plt.title("Hola")
plt.xticks(np.arange(12)+0.5,[f"Hola{i}" for i in range(12)], rotation=90)

plt.subplot(1,2,2)
sns.heatmap(uniform_data)
```
