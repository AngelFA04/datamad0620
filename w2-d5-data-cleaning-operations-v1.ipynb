```python
def fuelType(x):
    if x in ["Premium","Regular","Gasoline or E85","Premium or E85","Gasoline or natural gas"]:
        return "Gasoline"
    elif x in ["Diesel","CNG"]:
        return "Diesel"
    elif x in ["Electricity","Premium and Electricity","Regular Gas and Electricity","Regular Gas or Electricity"]:
        return "Hybrid"
    
    return "Others"


co2_cars["fuel_label"] = co2_cars.fuelType.apply(fuelType)
```

```python
co2_cars[co2_cars.fuel_label == "Hybrid"].groupby(["year","make"]).agg({"make":"count"})
```

```python
def change(q):
    return f"make:{q.make} model:{q.model}"

co2_cars["title"] = co2_cars.apply(change, axis=1)
```
