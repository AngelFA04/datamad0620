# ¿Qué es un servidor?

* ❌ Un sitio donde se guardan datos (hdd)
* ⚛️ Algo conectado a una red
* ✅ Cualquier ordenador puede tener un servidor

* La nube!!¿¿?¿?¿?

# UN SERVIDOR ES UN PROGRAMA

* url:  `https://www.marca.com/motor.html?intcmp=MENUPROD&s_kw=motor`

* Protocol: `https://`
* TLD: Top level domain `*.com`
* Domain Name: `marca.com`
* Subdomain: `www.marca.com`
* FQDN (fully qualified domain name): `marca.com` or `www.marca.com`
    
    
* IP: 127.0.0.1 (localhost) - 0.0.0.0 - 255.255.255.255

* Servidor DNS: Traduce nombres de dominio a IP: `dig www.marca.com`

* Endpoint: `/motor.html?intcmp=MENUPROD&s_kw=motor`
* Path: `/motor.html`
* Query parameters: `?intcmp=MENUPROD&s_kw=motor`

# JSON: recibiendo info de una api

```python
alumno = {
    "nombre":"Pepe",
    "edad":25,
    "notas":[10,10,10,6],
    "amigos":[
        {"nombre":"Juan"},
        {"nombre":"Marta"}
    ]
}
print(alumno)
```

```python
alumno_json = """
{
    "nombre":"Pepe",
    "edad":25,
    "notas":[10,10,10,6],
    "amigos":[
        {"nombre":"Juan"},
        {"nombre":"Marta"}
    ]
}
"""
print(alumno_json)
```

```python
print(type(alumno), type(alumno_json))
```

```python
alumno["nombre"]
```

```python
alumno_json["nombre"]
```

```python
import json
```

```python
# PARSEAR o DESERIALIZAR
datos = json.loads(alumno_json)
print(datos)
print(type(datos), type(alumno_json))
```

```python
import os
from dotenv import load_dotenv
load_dotenv()

def getFromGithub(endpoint, queryParams={}):
    url = f"https://api.github.com{endpoint}"

    apiKey = os.getenv("GITHUB_APIKEY")
    if not apiKey:
        raise ValueError("Please setup the GITHUB_APIKEY env variable")
    headers = {
        "Authorization": f"token {apiKey}"
    }
    
    res = requests.get(url, headers=headers, params=queryParams)
    print(res.url)
    return res.json()
```
