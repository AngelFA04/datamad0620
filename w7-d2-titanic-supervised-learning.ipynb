# Pipeline for training

```python
from sklearn.preprocessing import StandardScaler, Normalizer
from sklearn.pipeline import make_pipeline

pipeline = [
    StandardScaler(),
    Normalizer()
]

tr = make_pipeline(*pipeline)
```

```python
X_ready = tr.fit_transform(X)
X_ready = pd.DataFrame(X_ready, columns=X.columns)
```

```python
X_ready.shape
```

```python
X_ready
```

```python
y
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X_ready, y, test_size=0.2)
```

```python
print(X_train.shape,X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(X_train, y_train)
clf.coef_
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
clf.predict_proba(X_test)
```

```python
print(y_pred.shape)
```

```python
evaluation = pd.DataFrame({
    "gt":y_test,
    "pred":y_pred
})
```

```python
evaluation
```

```python
incorrectos = np.abs(evaluation["gt"]-evaluation["pred"]).sum()

incorrectos
```

```python
accuracy = 1-(incorrectos/len(evaluation["gt"]))
accuracy 
```
